# Question 1 (40 points)

Write a python program to collect text data from **either of the following sources** and save the data into a **csv file:**

(1) Collect all the customer reviews of a product (you can choose any porduct) on amazon. [atleast 1000 reviews]

(5) Collect all the information of the 904 narrators in the Densho Digital Repository.


```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re

def nar():
  rev=[]
  s=904
  HEADERS = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:98.0) Gecko/20100101 Firefox/98.0",
        "Accept": "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,/;q=0.8",
        "Accept-Language": "en-US,en;q=0.5",
        "Accept-Encoding": "gzip, deflate",
        "Connection": "keep-alive",
        "Upgrade-Insecure-Requests": "1",
        "Sec-Fetch-Dest": "document",
        "Sec-Fetch-Mode": "navigate",
        "Sec-Fetch-Site": "none",
        "Sec-Fetch-User": "?1",
        "Cache-Control": "max-age=0",
    }
  while len(rev)<s:
    url="https://ddr.densho.org/narrators/"
    r=requests.get(url,headers=HEADERS)
    #print(r.status_code)
    if r.status_code==200:
      soup=BeautifulSoup(r.content,"html.parser")
      res=soup.find_all("div",class_="media-body")
      for i in res:
        x={}
        name=i.find("h4",class_="media-heading").text.strip()
        info=i.find("div",class_="source muted").text.strip()
        if info:
          inf=info.split()
          date=None
          #loc=None
          if "Born" in inf  and "in" in inf:
            ib=inf.index("Born")
            ii=inf.index("in")
            date=" ".join(inf[ib+1:ii])
          """if "in" in inf and "." in inf:
            iii=inf.index("in")
            ip=inf.index(".")
            loc=" ".join(inf[iii+1:ip])"""
          rev.append([name,date,info])
        if len(rev)>=s:
          break
  return rev

data=nar()
df=pd.DataFrame(data,columns=["Name","Date Of Birth","Information"])
df.to_csv("nar",index=False)
print(df.to_string(index=False))
```

# Question 2 (30 points)

Write a python program to **clean the text data** you collected in the previous question and save the clean data in a new column in the csv file. The data cleaning steps include: [Code and output is required for each part]

(1) Remove noise, such as special characters and punctuations.

(2) Remove numbers.

(3) Remove stopwords by using the stopwords list.

(4) Lowercase all texts

(5) Stemming.

(6) Lemmatization.

```python
import pandas as pd
import string
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
nltk.download('stopwords')
nltk.download('punkt')

def clean(dat):
  dat=re.sub(r'[^a-zA-Z\s/]','',dat)
  dat=re.sub(r'\d+','',dat)
  dat=dat.lower()
  da=nltk.word_tokenize(dat)
  da=[i for i in da if i not in sw]
  da=[ps.stem(i) for i in da]
  da=[wl.lemmatize(i) for i in da]
  ct=' '.join(da)

  return ct

def main():
  df.to_csv("nar.csv")
  x=df['Information']
  y=x.apply(clean)
  df["Information"]=y
  df.to_csv("clean",index=False)
  print(df.to_string(index=False))



if __name__== "__main__":
  sw=set(stopwords.words('english'))
  ps=PorterStemmer()
  wl=WordNetLemmatizer()

  main()
```

# Question 3 (30 points)

Write a python program to **conduct syntax and structure analysis of the clean text** you just saved above. The syntax and structure analysis includes:

(1) **Parts of Speech (POS) Tagging:** Tag Parts of Speech of each word in the text, and calculate the total number of N(oun), V(erb), Adj(ective), Adv(erb), respectively.

(2) **Constituency Parsing and Dependency Parsing:** print out the constituency parsing trees and dependency parsing trees of all the sentences. Using one sentence as an example to explain your understanding about the constituency parsing tree and dependency parsing tree.

(3) **Named Entity Recognition:** Extract all the entities such as person names, organizations, locations, product names, and date from the clean texts, calculate the count of each entity.

```python
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.tag import pos_tag
from nltk.chunk import ne_chunk
import spacy

nlp = spacy.load("en_core_web_sm")

def pos_tagging(text):
    sentences = sent_tokenize(text)
    noun_count = 0
    verb_count = 0
    adj_count = 0
    adv_count = 0
    for sentence in sentences:
      words = word_tokenize(sentence)
      tagged_words = pos_tag(words)
      for word, tag in tagged_words:
        if tag.startswith('N'):
          noun_count += 1
        elif tag.startswith('V'):
          verb_count += 1
        elif tag.startswith('J'):
          adj_count += 1
        elif tag.startswith('R'):
          adv_count += 1
    return noun_count, verb_count, adj_count, adv_count

def constituency_dependency_parsing(text):
    doc = nlp(text)
    for sent in doc.sents:
        print("Sentence:", sent.text)
        print("Dependency Parsing Tree:")
        for token in sent:
            print(token.text, "-->", token.dep_, "-->", token.head.text)
        print()

def named_entity_recognition(text):
    doc = nlp(text)
    person_count = 0
    org_count = 0
    loc_count = 0
    product_count = 0
    date_count = 0
    for ent in doc.ents:
      if ent.label_ == "PERSON":
        person_count += 1
      elif ent.label_ == "ORG":
        org_count += 1
      elif ent.label_ == "LOC":
        loc_count += 1
      elif ent.label_ == "PRODUCT":
        product_count += 1
      elif ent.label_ == "DATE":
        date_count += 1
    return person_count, org_count, loc_count, product_count, date_count
def main():

  df.to_csv("clean.csv")
  ys=df['Information']
  ys=''.join(df['Information'].astype(str))

  clean_text=ys
  noun_count, verb_count, adj_count, adv_count = pos_tagging(clean_text)
  a1 = {"Parts of Speech": ["Noun", "Verb", "Adjective", "Adverb"],"Count": [noun_count, verb_count, adj_count, adv_count]}
  df1=pd.DataFrame(a1)

  person_count, org_count, loc_count, product_count, date_count = named_entity_recognition(clean_text)
  print("Named Entity Recognition:")
  a3= {"Entity": ["Person", "Organization", "Location", "Product", "Date"],"Count": [person_count, org_count, loc_count, product_count, date_count]}
  df2=pd.DataFrame(a3)
  print("Parts of Speech (POS) Tagging:")
  print(df1)
  print()
  print("Constituency and Dependency Parsing:")
  ct=df['Information'].iloc[0]
  constituency_dependency_parsing(ct)

  print()
  print("Named Entity Recognition:")
  print(df2)
if __name__== "__main__":
  main()
```
